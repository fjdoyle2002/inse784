<h1>This notebook looks at the <u>Central Limit Theorem</u> in respect to the sampling distribution of the mean</h1>
<h2>We are going to treat the set of integers from 1 to 99 as our population and take random samples of these from size n=1 to size n=30 to calculate the sample mean and see how the resulting distributions compare.</h2>

```python
import random
#This method will return the average of n integer samples pick at random between 
#1 and 99. The argument "sample_size" is the value for n
def get_sample_average(sample_size):
    sum = 0
    for x in range(sample_size):
        sum = sum + random.randint(1,99)
    avg = sum/sample_size
    return avg
```

```python
import pandas as pd
#we are going to store each set of samples for size "n" in columns of a dataframe
df = pd.DataFrame()
#we will create a sample set for each value of n from 1 to 30 (range(1,31) gives a set of int 
#values from 1 "up to, but not including" 31)
for n in range(1,31):
    name = "n = "+str(n)
    samp_list = []
    #we are going to collect the mean (i.e., average) for 1000 samples of size n
    #and store them in a list that will become the next column in the dataframe
    for sample in range(1000):
        samp_list.append(get_sample_average(n))
    df[name] = samp_list        
    
```

```python
df.describe()
```

```python
import seaborn as sns
sns.histplot(df["n = 1"], kde=True)
```

```python
##probability plot to help assess normality
import matplotlib.pyplot as plt
import scipy
fig, ax = plt.subplots(figsize=(3, 3))
scipy.stats.probplot(df['n = 1'], plot=ax)
```

```python
#remember, this test produce a p-val for a null hypothesis that the population IS normal. A small value for p means that 
#we should reject the null hypothesis and accept the alternative hypothesis that the population is NOT normal.
scipy.stats.normaltest(df['n = 1'])
```
