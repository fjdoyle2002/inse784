<h1>This notebook will introduce descriptive statistics concepts</h1>
<h2>Particularly:</h2>
<h3> Frequency Distributions (Histograms)</h3> 
<h3> Measures of Central Tendency (Mean/Median/Mode)</h3> 
<h3> Dispersion/Variation</h3>


```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./Salaries.csv")
sns.histplot(df['salary'], kde=False)
```

```python
#describe() can be used on a column as well as a whole dataframe and provides many 
#descriptive stats of interest including the mean, standard deviation and quartiles 
#remember, the median is the 50th percentile
df['salary'].describe()
```

```python
#note that return of mode() method not a single value but a series 
#data structure that could contain more than one value (because there can be multiple modes)
df['salary'].mode()
```

```python
#skew is a ma=easure of how assymetric the set is about its mean
df['salary'].skew()
```

```python
#kurtosis is often described as "peakedness" which describes the shape of the peak of the distribution as
#pointier or flatter. This is not really accurate. Kurtosis should be seen as a measure of how much of the data
#consists of outliers. It is more a measure of the tails of the distribution.
df['salary'].kurtosis()
```

<h2>Let's look at some relevant plots</h2>

```python
sns.boxplot(x="rank",y="salary",data = df)
```

```python
#violin plot gives us the information in a boxplot plus a density 
#plot with information analogous to the histogram
sns.set(style = 'whitegrid')  
sns.violinplot(x ="rank", y ="salary", data = df)
```

```python
#correlation tells us how multiple variables change in relation to each other
#hear, we plot correlation coefficients in a heatmap for easier visual detection of differences
sns.heatmap(df.corr(),annot=True,cmap="RdYlGn")
```

```python
#some more exploratory data analysis on full dataset using pairwise plots in seaborn
#we are creating a new dataframe and converting the 0/1 outcome value to a new column "out_cat"
#for "outcome category" with a text label of positive or negative and then using it to color the data points to group values
df2 = df.copy()
df2['out_cat'] = 'positive'
df2.loc[df2['outcome'] == 0, 'out_cat'] = 'negative'
df2 = df2.drop(columns="outcome")
sns.pairplot(df2,hue="out_cat", palette="Dark2")
```
